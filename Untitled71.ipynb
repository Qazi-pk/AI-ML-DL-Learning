````python
from pysr import PySRRegressor

print('Starting symbolic regression of V_eff(h) with PySR...')

X = h_range.cpu().numpy()  # Input features (h)
y = v_eff_pred.flatten()   # Target (V_eff)

# Initialize PySR regressor
# Common operators: +, -, *, /, cos, sin, exp, log, sqrt, abs, ^
# Note: exp and log can be tricky with negative values or small values near zero
model_pysr = PySRRegressor(
    select_k_features=1, # Corrected argument from n_features
    binary_operators=["+", "-", "*", "/", "^"],
    unary_operators=[
        "cos", "sin", "exp", "log",
        "sqrt", # sqrt of negative values is NaN, need to be careful
        ("neg", lambda x: -x)
    ],
    # Restrict operators to prevent issues with non-positive values
    # For instance, ensure log and sqrt don't get negative inputs
    constraints={
        'log': (1, 0), # Only 1 argument, but 0 output (no constraint on output)
        'sqrt': (1, 0) # Only 1 argument
    },
    # Optionally, specify specific functions to avoid problems
    # e.g., 'log_abs(x) = log(abs(x))' for log
    # Or use custom operators
    # e.g., "f_log(x) = log(abs(x)+1e-6)"

    # You might want to adjust populations, n_iterations, etc.
    populations=16,          # Number of populations to evolve
    n_iterations=5000,       # Number of iterations to run evolutionary search
    maxsize=20,              # Max size of an expression
    timeout_in_seconds=120,  # Max time for symbolic regression

    # Output options
    output_jax_format=False, # We want human-readable string
    output_latex=True,       # For LaTeX output

    # Verbosity
    verbosity=0,

    # Other parameters
    model_selection='best',
    procs=4 # Number of processes to use
)

# Fit the model
# PySR expects 2D input for X, 1D for y
model_pysr.fit(X, y)

print('Symbolic regression complete!')
print('\nTop expressions discovered:')

# Print the best equation
print(model_pysr.get_best()
)
# If you want to see all equations:
# print(model_pysr.equations)
````

````python
print('Preparing data for Neural Lagrangian: computing derivative...')

# Convert TimeSeries to numpy array for easier manipulation
h = filtered_strain.value
t = filtered_strain.times.value

# Calculate time step (assuming uniform sampling)
dt = t[1] - t[0]

# Compute numerical derivative using central difference
# ḣ(t) ≈ (h(t+Δt) - h(t-Δt)) / (2Δt)
# For boundary points, forward/backward difference might be needed,
# or simply trim the first/last point.

# Use np.gradient for a more robust numerical derivative handling boundaries
h_dot = np.gradient(h, dt)

print('✓ Derivative computed successfully!')
print(f'  Shape of h: {h.shape}')
print(f'  Shape of h_dot: {h_dot.shape}')

# Convert to PyTorch tensors
h_tensor = torch.tensor(h, dtype=torch.float32).to(device).unsqueeze(1)
h_dot_tensor = torch.tensor(h_dot, dtype=torch.float32).to(device).unsqueeze(1)

print(f'  h_tensor shape: {h_tensor.shape}')
print(f'  h_dot_tensor shape: {h_dot_tensor.shape}')
````

## Step 6: Define Neural Lagrangian Model

We define a neural network to represent the Lagrangian $L(h, \dot{h})$.

The Lagrangian is assumed to be of the form $L(h, \dot{h}) = \frac{1}{2} m \dot{h}^2 - V_{eff}(h)$, where $m$ is some effective mass (which can be absorbed into the network or learned), and $V_{eff}(h)$ is an effective potential. The neural network will learn $V_{eff}(h)$.

For a simpler approach initially, we can model the entire Lagrangian $L(h, \dot{h})$ directly with a neural network that takes both $h$ and $\dot{h}$ as inputs. However, to explicitly discover an effective potential, it's better to structure the network this way.

Here, we represent $V_{eff}(h)$ as a Multi-Layer Perceptron (MLP).

````python
class NeuralPotential(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(NeuralPotential, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, output_dim)
        )

    def forward(self, h):
        return self.network(h)

# Define Neural Lagrangian
# L = 0.5 * effective_mass * h_dot^2 - V_eff(h)
class NeuralLagrangian(nn.Module):
    def __init__(self, potential_net, effective_mass_init=1.0):
        super(NeuralLagrangian, self).__init__()
        self.potential_net = potential_net
        # Learnable effective mass parameter
        self.effective_mass = nn.Parameter(torch.tensor(effective_mass_init, dtype=torch.float32))

    def forward(self, h, h_dot):
        # Ensure effective_mass is positive
        m = torch.abs(self.effective_mass)
        kinetic_energy = 0.5 * m * h_dot**2
        potential_energy = self.potential_net(h)
        return kinetic_energy - potential_energy

input_dim = 1  # h is a scalar
hidden_dim = 64
output_dim = 1 # V_eff(h) is a scalar

potential_net = NeuralPotential(input_dim, hidden_dim, output_dim).to(device)
model = NeuralLagrangian(potential_net).to(device)

print('Neural Lagrangian model defined:')
print(model)
print(f'Initial effective mass: {model.effective_mass.item():.4f}')
````

## Step 7: Define Euler-Lagrange Residual

For a system described by a Lagrangian $L(q, \dot{q})$, the equations of motion are given by the Euler-Lagrange equation:

$\frac{\partial L}{\partial q} - \frac{d}{dt}\left(\frac{\partial L}{\partial \dot{q}}\right) = 0$

In our case, $q = h$. So, the equation becomes:

$\frac{\partial L}{\partial h} - \frac{d}{dt}\left(\frac{\partial L}{\partial \dot{h}}\right) = 0$

We define the Euler-Lagrange residual as the left-hand side of this equation. Our goal is to minimize this residual to find the Lagrangian that best describes the observed data.

Given $L(h, \dot{h}) = \frac{1}{2} m \dot{h}^2 - V_{eff}(h)$:

1. $\frac{\partial L}{\partial h} = -\frac{\partial V_{eff}}{\partial h}$

2. $\frac{\partial L}{\partial \dot{h}} = m \dot{h}$

3. $\frac{d}{dt}\left(\frac{\partial L}{\partial \dot{h}}\right) = \frac{d}{dt}(m \dot{h}) = m \ddot{h}$
   (Assuming $m$ is constant or slowly varying and learned as a scalar parameter)

So, the Euler-Lagrange residual is: $- \frac{\partial V_{eff}}{\partial h} - m \ddot{h}$

To compute $\frac{\partial V_{eff}}{\partial h}$, we'll use automatic differentiation. We also need $\ddot{h}$, which can be computed by taking the derivative of $\dot{h}$ (or second derivative of $h$).

````python
def euler_lagrange_residual(model, h, h_dot, dt):
    # Enable gradient computation for h for dV_eff/dh
    h.requires_grad_(True)

    # Compute L(h, h_dot)
    L = model(h, h_dot)

    # 1. Compute dL/dh = -dV_eff/dh
    # Sum L over all samples for scalar output for autograd
    dL_dh = torch.autograd.grad(L.sum(), h, create_graph=True)[0]

    # 2. Compute dL/dh_dot = m * h_dot
    # This is a direct computation from the model structure
    dL_dh_dot = torch.abs(model.effective_mass) * h_dot

    # 3. Compute d/dt(dL/dh_dot) = m * h_double_dot
    # Numerically differentiate dL_dh_dot with respect to time
    dL_dh_dot_np = dL_dh_dot.detach().cpu().numpy().flatten()
    ddt_dL_dh_dot_np = np.gradient(dL_dh_dot_np, dt)
    ddt_dL_dh_dot = torch.tensor(ddt_dL_dh_dot_np, dtype=torch.float32).to(device).unsqueeze(1)

    # Euler-Lagrange residual
    residual = dL_dh - ddt_dL_dh_dot

    # Reset requires_grad for h if it was set True locally
    h.requires_grad_(False)

    return residual
````

## Step 8: Training the Neural Lagrangian

We train the neural network by minimizing the squared Euler-Lagrange residual. The loss function will be the mean squared error of the residual.

````python
epochs = 500
learning_rate = 1e-3
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# Calculate dt from the original time series for numerical derivative
dt = filtered_strain.times.value[1] - filtered_strain.times.value[0]

losses = []

print('Starting training of Neural Lagrangian...')
for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()

    # Compute residual
    res = euler_lagrange_residual(model, h_tensor, h_dot_tensor, dt)

    # Loss is the mean squared residual
    loss = torch.mean(res**2)

    loss.backward()
    optimizer.step()

    losses.append(loss.item())

    if (epoch + 1) % 50 == 0:
        print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item():.8f}, Effective Mass: {torch.abs(model.effective_mass).item():.4f}')

print('Training complete!')

# Plot loss curve
plt.figure(figsize=(10, 6))
plt.plot(losses)
plt.xlabel('Epoch')
plt.ylabel('Loss (Mean Squared Euler-Lagrange Residual)')
plt.title('Neural Lagrangian Training Loss')
plt.grid(True)
plt.show()
````

## Step 9: Extract Effective Potential $V_{eff}(h)$

After training, the `potential_net` within our `NeuralLagrangian` model represents the effective potential $V_{eff}(h)$.

````python
model.eval()

# Create a range of h values to evaluate V_eff
h_min = h_tensor.min().item()
h_max = h_tensor.max().item()
h_range = torch.linspace(h_min, h_max, 500).to(device).unsqueeze(1)

# Predict V_eff for the range of h values
with torch.no_grad():
    v_eff_pred = model.potential_net(h_range).cpu().numpy()

# Plot the discovered effective potential
plt.figure(figsize=(10, 6))
plt.plot(h_range.cpu().numpy(), v_eff_pred)
plt.xlabel('Strain (h)')
plt.ylabel('Effective Potential ($V_{eff}(h)$)')
plt.title('Discovered Effective Potential for GW150914')
plt.grid(True)
plt.show()

print(f'Learned effective mass: {torch.abs(model.effective_mass).item():.4f}')
````

## Step 12: Final Discovered Action Principle

Based on the discovered effective potential from PySR, we can now write down the effective action principle for GW150914 data.

Recall that the Lagrangian is $L(h, \dot{h}) = \frac{1}{2} m \dot{h}^2 - V_{eff}(h)$.

The action $S = \int L(h, \dot{h}) dt$. Therefore, the discovered action principle is:

$S = \int \left( \frac{1}{2} m \dot{h}^2 - V_{eff}(h) \right) dt$
