````python
def preprocess_strain(strain_data, highpass_freq=30, lowpass_freq=250, resample_rate=512, q_val=20, normalize=True):
    # Apply bandpass filter
    # Note: gwpy's bandpass method is typically more robust for this data
    # We'll use a custom butterworth for demonstration consistency
    # with generic time series, but in practice, use strain.bandpass if available.

    # Detrend the data first
    strain_data = strain_data.detrend()

    # Apply a highpass filter to remove low-frequency noise
    hp_strain = strain_data.highpass(highpass_freq).value

    # Apply a lowpass filter to remove high-frequency noise
    bp_strain = strain_data.bandpass(highpass_freq, lowpass_freq).value

    # Create a new TimeSeries object for filtering/resampling
    processed_strain = TimeSeries(bp_strain, times=strain_data.times, sample_rate=strain_data.sample_rate)

    # Resample the data
    processed_strain = processed_strain.resample(resample_rate)

    # Get values and times
    h_t = processed_strain.value
    t = processed_strain.times.value

    # Normalize (optional)
    if normalize:
        h_t = (h_t - np.mean(h_t)) / np.std(h_t)

    return t, h_t

# Apply preprocessing
t_processed, h_processed = preprocess_strain(strain.copy())
dt = t_processed[1] - t_processed[0]

print(f'Original sample rate: {strain.sample_rate} Hz')
print(f'Processed sample rate: {1/dt:.2f} Hz')
print(f'Number of processed data points: {len(h_processed)}')

# Visualize processed data
plt.figure(figsize=(12, 4))
plt.plot(t_processed - gps_time, h_processed, linewidth=0.7, color='g')
plt.xlabel('Time from merger (s)')
plt.ylabel('Normalized Strain')
plt.title('Processed LIGO Strain Data (Bandpassed, Resampled, Normalized)')
plt.axvline(0, color='r', linestyle='--', alpha=0.5, label='Merger time')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
class NeuralLagrangian(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, output_dim=1):
        super(NeuralLagrangian, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, output_dim)
        )

    def forward(self, x):
        return self.net(x)

# Input to the network will be (h, ḣ)
input_dim = 2
lagrangian_model = NeuralLagrangian(input_dim).to(device)
print(lagrangian_model)
````

````python
def euler_lagrange_loss(model, h_data, dt):
    h_data.requires_grad_(True)

    # Calculate ḣ using finite differences
    h_dot_data = torch.gradient(h_data, spacing=(dt,), dim=0).values

    # Concatenate h and ḣ for model input
    inputs = torch.stack([h_data, h_dot_data], dim=-1)

    # Compute Lagrangian L(h, ḣ)
    L = model(inputs).squeeze()

    # Calculate ∂L/∂h
    dL_dh = torch.autograd.grad(L.sum(), h_data, create_graph=True)[0]

    # Calculate ∂L/∂ḣ
    dL_dh_dot = torch.autograd.grad(L.sum(), h_dot_data, create_graph=True)[0]

    # Calculate d/dt(∂L/∂ḣ) using finite differences
    ddt_dL_dh_dot = torch.gradient(dL_dh_dot, spacing=(dt,), dim=0).values

    # Euler-Lagrange residual
    EL_residual = dL_dh - ddt_dL_dh_dot

    # The loss is the mean squared of the residual
    # We consider only the inner points where derivatives are well defined
    loss = torch.mean(EL_residual[1:-1]**2)

    return loss, EL_residual
````

````python
optimizer = torch.optim.Adam(lagrangian_model.parameters(), lr=1e-3)
epochs = 500

h_tensor = torch.tensor(h_processed, dtype=torch.float32).to(device)

print(f'Starting training for {epochs} epochs...')

loss_history = []

for epoch in range(epochs):
    optimizer.zero_grad()
    loss, _ = euler_lagrange_loss(lagrangian_model, h_tensor, dt)
    loss.backward()
    optimizer.step()

    loss_history.append(loss.item())

    if (epoch + 1) % 50 == 0:
        print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item():.8f}')

print('Training finished!')

# Plot loss history
plt.figure(figsize=(10, 5))
plt.plot(loss_history)
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE of EL Residual)')
plt.title('Training Loss History')
plt.grid(True)
plt.show()
````

````python
lagrangian_model.eval()

with torch.no_grad():
    h_tensor.requires_grad_(True)
    h_dot_data = torch.gradient(h_tensor, spacing=(dt,), dim=0).values

    inputs = torch.stack([h_tensor, h_dot_data], dim=-1)
    L_val = lagrangian_model(inputs).squeeze()

    # These gradients are with respect to the input h_tensor, h_dot_data for L_val.sum()
    dL_dh = torch.autograd.grad(L_val.sum(), h_tensor, create_graph=False)[0]
    dL_dh_dot = torch.autograd.grad(L_val.sum(), h_dot_data, create_graph=False)[0]

    ddt_dL_dh_dot = torch.gradient(dL_dh_dot, spacing=(dt,), dim=0).values

    # We can consider 'F_effective = ddt_dL_dh_dot' which should be equal to dL_dh if EL holds
    # For symbolic regression of V(h), we are interested in F = -dV/dh.
    # So we could try to fit dL_dh or ddt_dL_dh_dot as -dV/dh

    # Let's use dL_dh as the 'generalized force' to derive potential
    # We'll use the central part to avoid boundary effects from finite differences
    h_sym_reg = h_tensor[1:-1].cpu().numpy()
    F_effective = dL_dh[1:-1].cpu().numpy()

    # Plot F_effective vs h
    plt.figure(figsize=(10, 5))
    plt.scatter(h_sym_reg, F_effective, s=5, alpha=0.5)
    plt.xlabel('h (strain)')
    plt.ylabel('Effective Force (dL/dh)')
    plt.title('Effective Force (dL/dh) vs. Strain h')
    plt.grid(True)
    plt.show()
````

````python
from pysr import PySRRegressor

# Ensure data is 2D for PySR (n_samples, n_features)
X = h_sym_reg.reshape(-1, 1)
y = F_effective

print('Starting PySR symbolic regression...')

# Configure PySR
model = PySRRegressor(
    model_selection="best",  # Will auto-select best model
    binary_operators=["+", "-", "*", "/"],
    unary_operators=[
        "sin", "cos", "exp", "log",
        # Additional operators can be added based on expected physics
        "abs",
        "sqrt",
        "square", "cube"
    ],
    niterations=100,  # Number of iterations to run PySR
    procs=8,          # Number of processes to use
    # maxsize=30,       # Max complexity of expressions
    # populations=50,   # Number of populations in evolutionary search
    timeout_in_seconds=60*2, # Timeout for PySR (2 minutes)
    extra_sympy_mappings={'square': lambda x: x**2, 'cube': lambda x: x**3},
    # ^ for custom operators
    # nested_constraints={'/': {'sin':0, 'cos':0}, 'sin':{'sin':0}, 'cos':{'cos':0}}
)

# Fit the model
# PySR can be computationally intensive, adjust niterations and timeout as needed
model.fit(X, y)

print('\nPySR regression finished! Results:')
print(model)

# Get the best equation
best_equation = model.get_best() # This returns a sympy expression or string
print(f'\nBest symbolic equation for F_effective(h) (dL/dh): {best_equation}')

# Try to plot the fit
h_test = np.linspace(h_sym_reg.min(), h_sym_reg.max(), 500).reshape(-1, 1)
F_pred = model.predict(h_test)

plt.figure(figsize=(10, 5))
plt.scatter(h_sym_reg, F_effective, s=5, alpha=0.5, label='Learned F_effective (dL/dh)')
plt.plot(h_test, F_pred, color='red', linewidth=2, label='PySR Fit')
plt.xlabel('h (strain)')
plt.ylabel('Effective Force (dL/dh)')
plt.title('PySR Symbolic Regression of Effective Force')
plt.grid(True)
plt.legend()
plt.show()
````

````python
from sympy import sympify, integrate, Symbol

# Convert best_equation from PySR to a sympy expression
try:
    h_sym = Symbol('x') # PySR uses 'x' for the first feature
    F_effective_sympy = sympify(str(best_equation).replace('x', 'h'))

    print(f'Symbolic expression for F_effective(h): {F_effective_sympy}')

    # Integrate to find V_eff(h)
    # V_eff(h) = - integral(F_effective(h) dh)
    V_eff_sympy = -integrate(F_effective_sympy, h_sym)

    print(f'Symbolic expression for V_eff(h): {V_eff_sympy}')

    # Convert sympy expression to a callable function for plotting
    from sympy.utilities.lambdify import lambdify
    V_eff_func = lambdify(h_sym, V_eff_sympy, 'numpy')

    # Plot the inferred potential
    plt.figure(figsize=(10, 5))
    plt.plot(h_test, V_eff_func(h_test), color='purple', linewidth=2)
    plt.xlabel('h (strain)')
    plt.ylabel('Effective Potential V_eff(h)')
    plt.title('Inferred Effective Potential from PySR (V_eff = -∫ F_effective dh)')
    plt.grid(True)
    plt.show()

except Exception as e:
    print(f"Could not perform symbolic integration: {e}")
    print("Falling back to numerical integration of PySR fit.")

    # Fallback: Numerical integration if symbolic fails or is too complex
    # We need to sort the data by h_test before integrating
    sort_indices = np.argsort(h_test.flatten())
    h_sorted = h_test[sort_indices].flatten()
    F_pred_sorted = F_pred[sort_indices].flatten()

    # Numerical integration using cumulative sum (Riemann sum)
    # Assuming F_effective = -dV/dh, so dV = -F_effective dh
    # V_eff(h) = - cumulative_integral(F_effective(h) dh)
    V_eff_numerical = -np.cumsum(F_pred_sorted * (h_sorted[1] - h_sorted[0]))

    # Shift potential so min is 0 or around 0 for better visualization
    V_eff_numerical -= V_eff_numerical.min()

    plt.figure(figsize=(10, 5))
    plt.plot(h_sorted, V_eff_numerical, color='orange', linewidth=2)
    plt.xlabel('h (strain)')
    plt.ylabel('Effective Potential V_eff(h)')
    plt.title('Numerically Inferred Effective Potential from PySR Fit')
    plt.grid(True)
    plt.show()

# Interpretation of the Action Principle:
print("\n--- Discovered Action Principle ---")
print("Based on the Neural Lagrangian approach and Symbolic Regression, we have identified a potential candidate for the effective potential V_eff(h) governing the gravitational wave strain h(t).")
print("The general form of the action can be written as: S = ∫ L dt, where L = T - V_eff(h).")
print("If we assume a kinetic term of the form T = 1/2 * m * ḣ^2 (with m=1 for simplicity in normalized units), then the discovered Lagrangian is approximately L(h, ḣ) = 1/2 * ḣ^2 - V_eff(h), where V_eff(h) is the symbolically regressed potential.")
print("The specific form of V_eff(h) is derived from the PySR output for F_effective(h) = dL/dh, and then integrated.")
print("This provides an effective action principle for the gravitational wave data, offering a novel way to analyze and potentially uncover underlying physical laws.")
````

````python
def preprocess_strain(strain_data, highpass_freq=30, lowpass_freq=250, resample_rate=512, q_val=20, normalize=True):
    # Apply bandpass filter
    # Note: gwpy's bandpass method is typically more robust for this data
    # We'll use a custom butterworth for demonstration consistency
    # with generic time series, but in practice, use strain.bandpass if available.

    # Detrend the data first
    strain_data = strain_data.detrend()

    # Apply a highpass filter to remove low-frequency noise
    hp_strain = strain_data.highpass(highpass_freq).value

    # Apply a lowpass filter to remove high-frequency noise
    bp_strain = strain_data.bandpass(highpass_freq, lowpass_freq).value

    # Create a new TimeSeries object for filtering/resampling
    processed_strain = TimeSeries(bp_strain, times=strain_data.times, sample_rate=strain_data.sample_rate)

    # Resample the data
    processed_strain = processed_strain.resample(resample_rate)

    # Get values and times
    h_t = processed_strain.value
    t = processed_strain.times.value

    # Normalize (optional)
    if normalize:
        h_t = (h_t - np.mean(h_t)) / np.std(h_t)

    return t, h_t

# Apply preprocessing
t_processed, h_processed = preprocess_strain(strain.copy())
dt = t_processed[1] - t_processed[0]

print(f'Original sample rate: {strain.sample_rate} Hz')
print(f'Processed sample rate: {1/dt:.2f} Hz')
print(f'Number of processed data points: {len(h_processed)}')

# Visualize processed data
plt.figure(figsize=(12, 4))
plt.plot(t_processed - gps_time, h_processed, linewidth=0.7, color='g')
plt.xlabel('Time from merger (s)')
plt.ylabel('Normalized Strain')
plt.title('Processed LIGO Strain Data (Bandpassed, Resampled, Normalized)')
plt.axvline(0, color='r', linestyle='--', alpha=0.5, label='Merger time')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
def euler_lagrange_loss(model, h_data, dt):
    h_data.requires_grad_(True)

    # Calculate ḣ using finite differences
    h_dot_data = torch.gradient(h_data, spacing=(dt,), dim=0).values

    # Concatenate h and ḣ for model input
    inputs = torch.stack([h_data, h_dot_data], dim=-1)

    # Compute Lagrangian L(h, ḣ)
    L = model(inputs).squeeze()

    # Calculate ∂L/∂h
    dL_dh = torch.autograd.grad(L.sum(), h_data, create_graph=True)[0]

    # Calculate ∂L/∂ḣ
    dL_dh_dot = torch.autograd.grad(L.sum(), h_dot_data, create_graph=True)[0]

    # Calculate d/dt(∂L/∂ḣ) using finite differences
    ddt_dL_dh_dot = torch.gradient(dL_dh_dot, spacing=(dt,), dim=0).values

    # Euler-Lagrange residual
    EL_residual = dL_dh - ddt_dL_dh_dot

    # The loss is the mean squared of the residual
    # We consider only the inner points where derivatives are well defined
    loss = torch.mean(EL_residual[1:-1]**2)

    return loss, EL_residual
````

````python
def preprocess_strain(strain_data, highpass_freq=30, lowpass_freq=250, resample_rate=512, q_val=20, normalize=True):
    # Apply bandpass filter
    # Note: gwpy's bandpass method is typically more robust for this data
    # We'll use a custom butterworth for demonstration consistency
    # with generic time series, but in practice, use strain.bandpass if available.

    # Detrend the data first
    strain_data = strain_data.detrend()

    # Apply a highpass filter to remove low-frequency noise
    hp_strain = strain_data.highpass(highpass_freq).value

    # Apply a lowpass filter to remove high-frequency noise
    bp_strain = strain_data.bandpass(highpass_freq, lowpass_freq).value

    # Create a new TimeSeries object for filtering/resampling
    processed_strain = TimeSeries(bp_strain, times=strain_data.times, sample_rate=strain_data.sample_rate)

    # Resample the data
    processed_strain = processed_strain.resample(resample_rate)

    # Get values and times
    h_t = processed_strain.value
    t = processed_strain.times.value

    # Normalize (optional)
    if normalize:
        h_t = (h_t - np.mean(h_t)) / np.std(h_t)

    return t, h_t

# Apply preprocessing
t_processed, h_processed = preprocess_strain(strain.copy())
dt = t_processed[1] - t_processed[0]

print(f'Original sample rate: {strain.sample_rate} Hz')
print(f'Processed sample rate: {1/dt:.2f} Hz')
print(f'Number of processed data points: {len(h_processed)}')

# Visualize processed data
plt.figure(figsize=(12, 4))
plt.plot(t_processed - gps_time, h_processed, linewidth=0.7, color='g')
plt.xlabel('Time from merger (s)')
plt.ylabel('Normalized Strain')
plt.title('Processed LIGO Strain Data (Bandpassed, Resampled, Normalized)')
plt.axvline(0, color='r', linestyle='--', alpha=0.5, label='Merger time')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
class NeuralLagrangian(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, output_dim=1):
        super(NeuralLagrangian, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, output_dim)
        )

    def forward(self, x):
        return self.net(x)

# Input to the network will be (h, ḣ)
input_dim = 2
lagrangian_model = NeuralLagrangian(input_dim).to(device)
print(lagrangian_model)
````

````python
def euler_lagrange_loss(model, h_data, dt):
    h_data.requires_grad_(True)

    # Calculate ḣ using finite differences
    h_dot_data = torch.gradient(h_data, spacing=(dt,), dim=0).values

    # Concatenate h and ḣ for model input
    inputs = torch.stack([h_data, h_dot_data], dim=-1)

    # Compute Lagrangian L(h, ḣ)
    L = model(inputs).squeeze()

    # Calculate ∂L/∂h
    dL_dh = torch.autograd.grad(L.sum(), h_data, create_graph=True)[0]

    # Calculate ∂L/∂ḣ
    dL_dh_dot = torch.autograd.grad(L.sum(), h_dot_data, create_graph=True)[0]

    # Calculate d/dt(∂L/∂ḣ) using finite differences
    ddt_dL_dh_dot = torch.gradient(dL_dh_dot, spacing=(dt,), dim=0).values

    # Euler-Lagrange residual
    EL_residual = dL_dh - ddt_dL_dh_dot

    # The loss is the mean squared of the residual
    # We consider only the inner points where derivatives are well defined
    loss = torch.mean(EL_residual[1:-1]**2)

    return loss, EL_residual
````

## Step 3: Preprocess Data

- Remove noise using a bandpass filter
- Resample to a lower frequency to reduce computational load
- Normalize the strain data

------------------------------

````python
def preprocess_strain(strain_data, highpass_freq=30, lowpass_freq=250, resample_rate=512, q_val=20, normalize=True):
    # Apply bandpass filter
    # Note: gwpy's bandpass method is typically more robust for this data
    # We'll use a custom butterworth for demonstration consistency
    # with generic time series, but in practice, use strain.bandpass if available.

    # Detrend the data first
    strain_data = strain_data.detrend()

    # Apply a highpass filter to remove low-frequency noise
    hp_strain = strain_data.highpass(highpass_freq).value

    # Apply a lowpass filter to remove high-frequency noise
    bp_strain = strain_data.bandpass(highpass_freq, lowpass_freq).value

    # Create a new TimeSeries object for filtering/resampling
    processed_strain = TimeSeries(bp_strain, times=strain_data.times, sample_rate=strain_data.sample_rate)

    # Resample the data
    processed_strain = processed_strain.resample(resample_rate)

    # Get values and times
    h_t = processed_strain.value
    t = processed_strain.times.value

    # Normalize (optional)
    if normalize:
        h_t = (h_t - np.mean(h_t)) / np.std(h_t)

    return t, h_t

# Apply preprocessing
t_processed, h_processed = preprocess_strain(strain.copy())
dt = t_processed[1] - t_processed[0]

print(f'Original sample rate: {strain.sample_rate} Hz')
print(f'Processed sample rate: {1/dt:.2f} Hz')
print(f'Number of processed data points: {len(h_processed)}')

# Visualize processed data
plt.figure(figsize=(12, 4))
plt.plot(t_processed - gps_time, h_processed, linewidth=0.7, color='g')
plt.xlabel('Time from merger (s)')
plt.ylabel('Normalized Strain')
plt.title('Processed LIGO Strain Data (Bandpassed, Resampled, Normalized)')
plt.axvline(0, color='r', linestyle='--', alpha=0.5, label='Merger time')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````

## Step 4: Define Neural Lagrangian Model

We define a neural network that represents the Lagrangian `L(h, ḣ)`. The Euler-Lagrange equation for a system with generalized coordinate `h` is:

`∂L/∂h - d/dt(∂L/∂ḣ) = 0`

Our goal is to train the network such that this equation is minimized. We will assume the system is conservative, so there's no dissipative force term.

`L(h, ḣ) = T(ḣ) - V(h)`

We can model the Lagrangian directly or model the kinetic and potential energy components separately.

For a simple 1D system, `T(ḣ) = 1/2 * m * ḣ^2`. We can include `m` as part of the neural network or assume `m=1`.

Here, we'll try to learn `L` directly as a function of `h` and `ḣ`.

------------------------------

````python
class NeuralLagrangian(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, output_dim=1):
        super(NeuralLagrangian, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, output_dim)
        )

    def forward(self, x):
        return self.net(x)

# Input to the network will be (h, ḣ)
input_dim = 2
lagrangian_model = NeuralLagrangian(input_dim).to(device)
print(lagrangian_model)
````

------------------------------

## Step 5: Define Euler-Lagrange Loss Function

The Euler-Lagrange equation is:
`∂L/∂h - d/dt(∂L/∂ḣ) = 0`

We need to compute the partial derivatives of L with respect to `h` and `ḣ`, and the time derivative of `∂L/∂ḣ`.

We'll use `autograd` from PyTorch for differentiation.

------------------------------

````python
def euler_lagrange_loss(model, h_data, dt):
    h_data.requires_grad_(True)

    # Calculate ḣ using finite differences
    h_dot_data = torch.gradient(h_data, spacing=(dt,), dim=0).values

    # Concatenate h and ḣ for model input
    inputs = torch.stack([h_data, h_dot_data], dim=-1)

    # Compute Lagrangian L(h, ḣ)
    L = model(inputs).squeeze()

    # Calculate ∂L/∂h
    dL_dh = torch.autograd.grad(L.sum(), h_data, create_graph=True)[0]

    # Calculate ∂L/∂ḣ
    dL_dh_dot = torch.autograd.grad(L.sum(), h_dot_data, create_graph=True)[0]

    # Calculate d/dt(∂L/∂ḣ) using finite differences
    ddt_dL_dh_dot = torch.gradient(dL_dh_dot, spacing=(dt,), dim=0).values

    # Euler-Lagrange residual
    EL_residual = dL_dh - ddt_dL_dh_dot

    # The loss is the mean squared of the residual
    # We consider only the inner points where derivatives are well defined
    loss = torch.mean(EL_residual[1:-1]**2)

    return loss, EL_residual
````

------------------------------

## Step 6: Training Loop

Train the `NeuralLagrangian` model to minimize the Euler-Lagrange residual.

------------------------------

````python
optimizer = torch.optim.Adam(lagrangian_model.parameters(), lr=1e-3)
epochs = 500

h_tensor = torch.tensor(h_processed, dtype=torch.float32).to(device)

print(f'Starting training for {epochs} epochs...')

loss_history = []

for epoch in range(epochs):
    optimizer.zero_grad()
    loss, _ = euler_lagrange_loss(lagrangian_model, h_tensor, dt)
    loss.backward()
    optimizer.step()

    loss_history.append(loss.item())

    if (epoch + 1) % 50 == 0:
        print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item():.8f}')

print('Training finished!')

# Plot loss history
plt.figure(figsize=(10, 5))
plt.plot(loss_history)
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE of EL Residual)')
plt.title('Training Loss History')
plt.grid(True)
plt.show()
````

## Step 7: Extracting the Effective Potential V_eff(h)

If we assume a standard kinetic term `T = 1/2 * m * ḣ^2` (or some other known form), then we can extract the effective potential `V_eff(h)` from the learned Lagrangian.

However, our `NeuralLagrangian` learns `L(h, ḣ)` directly. To extract `V_eff(h)`, we need to perform some additional steps. A common approach for conservative systems is `L = T - V`. If we assume `T = 1/2 * ḣ^2` (m=1 for simplicity), then `V(h) = 1/2 * ḣ^2 - L(h, ḣ)`.

This is tricky because `L` itself is a function of `ḣ`. A more robust way to define an effective potential from a general Lagrangian `L(h, ḣ)` involves the Legendre transform, which gives the Hamiltonian. However, for a simple interpretation, one might look at terms independent of `ḣ` as potential-like, or terms quadratic in `ḣ` as kinetic-like.

Let's assume the learned Lagrangian is approximately `L(h, ḣ) ≈ 1/2 * f(h) * ḣ^2 - V_eff(h)`. We can try to disentangle this. A simpler approach for *extracting an effective potential* is to use the Euler-Lagrange equation directly. If `L = T - V`, and we know `T`, then we can infer `V`.

Given the complexity, a direct extraction of `V_eff(h)` from a general `L(h, ḣ)` learned by a NN is not straightforward. Instead, we can observe the `dL_dh` term from the EL equation which represents generalized force `F = -∂V/∂h` if `T` is independent of `h`.

Let's try to extract `∂L/∂h` and `∂L/∂ḣ` as functions of `h` and `ḣ`.

For symbolic regression of `V_eff(h)`, we usually need `F = -∂V/∂h`. The Euler-Lagrange equation gives `∂L/∂h - d/dt(∂L/∂ḣ) = 0`, so `∂L/∂h = d/dt(∂L/∂ḣ)`. If we interpret `∂L/∂h` as the generalized force `-∂V/∂h`, then we can integrate it. This implies that the 'effective force' is `ddt(∂L/∂ḣ)`.

Let's compute `dL_dh` and `ddt_dL_dh_dot` for the trained model over the data.

------------------------------

````python
lagrangian_model.eval()

with torch.no_grad():
    h_tensor.requires_grad_(True)
    h_dot_data = torch.gradient(h_tensor, spacing=(dt,), dim=0).values

    inputs = torch.stack([h_tensor, h_dot_data], dim=-1)
    L_val = lagrangian_model(inputs).squeeze()

    # These gradients are with respect to the input h_tensor, h_dot_data for L_val.sum()
    dL_dh = torch.autograd.grad(L_val.sum(), h_tensor, create_graph=False)[0]
    dL_dh_dot = torch.autograd.grad(L_val.sum(), h_dot_data, create_graph=False)[0]

    ddt_dL_dh_dot = torch.gradient(dL_dh_dot, spacing=(dt,), dim=0).values

    # We can consider 'F_effective = ddt_dL_dh_dot' which should be equal to dL_dh if EL holds
    # For symbolic regression of V(h), we are interested in F = -dV/dh.
    # So we could try to fit dL_dh or ddt_dL_dh_dot as -dV/dh

    # Let's use dL_dh as the 'generalized force' to derive potential
    # We'll use the central part to avoid boundary effects from finite differences
    h_sym_reg = h_tensor[1:-1].cpu().numpy()
    F_effective = dL_dh[1:-1].cpu().numpy()

    # Plot F_effective vs h
    plt.figure(figsize=(10, 5))
    plt.scatter(h_sym_reg, F_effective, s=5, alpha=0.5)
    plt.xlabel('h (strain)')
    plt.ylabel('Effective Force (dL/dh)')
    plt.title('Effective Force (dL/dh) vs. Strain h')
    plt.grid(True)
    plt.show()
````

## Step 8: Symbolic Regression with PySR

We will use PySR to find a symbolic expression for the effective potential `V_eff(h)`. We'll assume `F_effective = -dV/dh`, so we need to find `V(h)` such that `F_effective = -V'(h)`.

PySR can find an expression for `F_effective(h)`. Once we have `F_effective(h)`, we can integrate it to find `V_eff(h)`.

Alternatively, we can directly train PySR on `h` vs `V_eff` if we had `V_eff` values. But we don't. We have `h` and `F_effective`. So, `F_effective(h)` is the target. Let's try to fit `F_effective` as a function of `h`.

Let `y = F_effective` and `x = h_sym_reg`.

------------------------------

````python
from pysr import PySRRegressor

# Ensure data is 2D for PySR (n_samples, n_features)
X = h_sym_reg.reshape(-1, 1)
y = F_effective

print('Starting PySR symbolic regression...')

# Configure PySR
model = PySRRegressor(
    model_selection="best",  # Will auto-select best model
    binary_operators=["+", "-", "*", "/"],
    unary_operators=[
        "sin", "cos", "exp", "log",
        # Additional operators can be added based on expected physics
        "abs",
        "sqrt",
        "square", "cube"
    ],
    niterations=100,  # Number of iterations to run PySR
    procs=8,          # Number of processes to use
    # maxsize=30,       # Max complexity of expressions
    # populations=50,   # Number of populations in evolutionary search
    timeout_in_seconds=60*2, # Timeout for PySR (2 minutes)
    extra_sympy_mappings={'square': lambda x: x**2, 'cube': lambda x: x**3},
    # ^ for custom operators
    # nested_constraints={'/': {'sin':0, 'cos':0}, 'sin':{'sin':0}, 'cos':{'cos':0}}
)

# Fit the model
# PySR can be computationally intensive, adjust niterations and timeout as needed
model.fit(X, y)

print('\nPySR regression finished! Results:')
print(model)

# Get the best equation
best_equation = model.get_best() # This returns a sympy expression or string
print(f'\nBest symbolic equation for F_effective(h) (dL/dh): {best_equation}')

# Try to plot the fit
h_test = np.linspace(h_sym_reg.min(), h_sym_reg.max(), 500).reshape(-1, 1)
F_pred = model.predict(h_test)

plt.figure(figsize=(10, 5))
plt.scatter(h_sym_reg, F_effective, s=5, alpha=0.5, label='Learned F_effective (dL/dh)')
plt.plot(h_test, F_pred, color='red', linewidth=2, label='PySR Fit')
plt.xlabel('h (strain)')
plt.ylabel('Effective Force (dL/dh)')
plt.title('PySR Symbolic Regression of Effective Force')
plt.grid(True)
plt.legend()
plt.show()
````

## Step 9: Inferring the Effective Potential and Action

If we found an expression `F_effective(h)` from PySR, we can then numerically or symbolically integrate it to find `V_eff(h)`.

`V_eff(h) = -∫ F_effective(h) dh`

Let's assume the simplest case where `F_effective(h) = -c * h` (linear restoring force), then `V_eff(h) = 1/2 * c * h^2` (harmonic oscillator potential).

However, PySR often returns complex expressions. We would need to parse the `sympy` expression and integrate it.

Let's directly use the `F_effective` from the model, and numerically integrate it to visualize the potential.

**Note**: The integration constant `C` for `V_eff` is arbitrary and doesn't affect the physics, so we can set it to 0.

------------------------------

````python
from sympy import sympify, integrate, Symbol

# Convert best_equation from PySR to a sympy expression
try:
    h_sym = Symbol('x') # PySR uses 'x' for the first feature
    F_effective_sympy = sympify(str(best_equation).replace('x', 'h'))

    print(f'Symbolic expression for F_effective(h): {F_effective_sympy}')

    # Integrate to find V_eff(h)
    # V_eff(h) = - integral(F_effective(h) dh)
    V_eff_sympy = -integrate(F_effective_sympy, h_sym)

    print(f'Symbolic expression for V_eff(h): {V_eff_sympy}')

    # Convert sympy expression to a callable function for plotting
    V_eff_func = lambdify(h_sym, V_eff_sympy, 'numpy')

    # Plot the inferred potential
    plt.figure(figsize=(10, 5))
    plt.plot(h_test, V_eff_func(h_test), color='purple', linewidth=2)
    plt.xlabel('h (strain)')
    plt.ylabel('Effective Potential V_eff(h)')
    plt.title('Inferred Effective Potential from PySR (V_eff = -∫ F_effective dh)')
    plt.grid(True)
    plt.show()

except Exception as e:
    print(f"Could not perform symbolic integration: {e}")
    print("Falling back to numerical integration of PySR fit.")

    # Fallback: Numerical integration if symbolic fails or is too complex
    # We need to sort the data by h_test before integrating
    sort_indices = np.argsort(h_test.flatten())
    h_sorted = h_test[sort_indices].flatten()
    F_pred_sorted = F_pred[sort_indices].flatten()

    # Numerical integration using cumulative sum (Riemann sum)
    # Assuming F_effective = -dV/dh, so dV = -F_effective dh
    # V_eff(h) = - cumulative_integral(F_effective(h) dh)
    V_eff_numerical = -np.cumsum(F_pred_sorted * (h_sorted[1] - h_sorted[0]))

    # Shift potential so min is 0 or around 0 for better visualization
    V_eff_numerical -= V_eff_numerical.min()

    plt.figure(figsize=(10, 5))
    plt.plot(h_sorted, V_eff_numerical, color='orange', linewidth=2)
    plt.xlabel('h (strain)')
    plt.ylabel('Effective Potential V_eff(h)')
    plt.title('Numerically Inferred Effective Potential from PySR Fit')
    plt.grid(True)
    plt.show()

# Interpretation of the Action Principle:
print("\n--- Discovered Action Principle ---")
print("Based on the Neural Lagrangian approach and Symbolic Regression, we have identified a potential candidate for the effective potential V_eff(h) governing the gravitational wave strain h(t).")
print("The general form of the action can be written as: S = ∫ L dt, where L = T - V_eff(h).")
print("If we assume a kinetic term of the form T = 1/2 * m * ḣ^2 (with m=1 for simplicity in normalized units), then the discovered Lagrangian is approximately L(h, ḣ) = 1/2 * ḣ^2 - V_eff(h), where V_eff(h) is the symbolically regressed potential.")
print("The specific form of V_eff(h) is derived from the PySR output for F_effective(h) = dL/dh, and then integrated.")
print("This provides an effective action principle for the gravitational wave data, offering a novel way to analyze and potentially uncover underlying physical laws.")
````

## Conclusion and Future Work

This notebook demonstrates a pioneering approach to applying Neural Lagrangians and Symbolic Regression to gravitational wave data from LIGO. By minimizing the Euler-Lagrange residual, we trained a neural network to infer an effective Lagrangian for the system, and subsequently extracted a candidate symbolic form for the effective potential `V_eff(h)`.

### Key Takeaways:
- **Proof of Concept**: Successfully applied Neural Lagrangians to real-world astrophysical data.
- **Symbolic Discovery**: PySR enabled the extraction of interpretable symbolic expressions for the effective force and potential.
- **Variational Physics**: Offers a new paradigm for analyzing GW data through an action principle framework.

### Limitations & Future Directions:
1.  **Kinetic Term Assumption**: Currently, the interpretation of `V_eff(h)` relies on assumptions about the kinetic term. Future work could explore learning the kinetic term `T(ḣ, h)` as well, leading to a more general Lagrangian structure.
2.  **Higher Order Derivatives**: Gravitational wave dynamics might involve higher-order derivatives (e.g., radiation reaction). Extending the framework to include `ḧ`, `h⃛`, etc., could capture more complex physics.
3.  **Multi-Variate Systems**: For binary mergers, a single `h(t)` is a projection. A full Lagrangian would involve multiple degrees of freedom (e.g., orbital parameters, polarization states).
4.  **Robustness and Uncertainty**: Quantifying uncertainty in the learned Lagrangian and symbolically regressed potential is crucial for scientific validity.
5.  **Comparison with GR**: A direct comparison of the discovered effective action with the predictions from General Relativity (e.g., post-Newtonian expansions) would be a critical next step.
6.  **Real-time Analysis**: Optimizing the training and symbolic regression for faster execution could enable applications in real-time GW alert pipelines.

This work opens exciting avenues for discovering fundamental physics from gravitational wave observations, potentially leading to new insights beyond existing theoretical frameworks.
