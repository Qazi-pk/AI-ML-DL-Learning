## Introduction and Background

Gravitational-wave (GW) observatories, such as the Laser Interferometer Gravitational-Wave Observatory (LIGO), measure distortions in spacetime caused by astrophysical events like merging black holes and neutron stars.


### The LIGO Interferometer

LIGO is a giant Michelson interferometer which splits laser light into two perpendicular arms, reflects it off mirrors at the end of each arm, and then recombines the light to produce an interference pattern. When a gravitational wave passes through the interferometer, it distorts spacetime, causing the arms to momentarily change length by a tiny amount. This change alters the phase difference between the light in the two arms, which in turn changes the interference pattern detected by the photodetector at the output of the interferometer. By precisely measuring these minute changes, LIGO can detect and characterize gravitational waves. For more information, refer to the [LIGO Website](https://www.ligo.caltech.edu/page/what-is-ligo).


### Problem Formulation

For this project, we are studying a single degree of freedom model of the mirror in one arm of the LIGO interferometer. We want to apply Lagrange's equations to understand the dynamics of the mirror, specifically how it responds to the forces acting upon it. The system consists of a mirror with mass $m$ attached to a spring with spring constant $k$ and a dashpot with damping coefficient $b$. An external force $F(t)$ acts on the mirror, which in our case is caused by the gravitational wave. The mirror's position is denoted by $x(t)$.


We start by defining the kinetic and potential energies of the system.

*   **Kinetic Energy ($T$):** The kinetic energy of the mirror is due to its motion and is given by $T = \frac{1}{2}m\dot{x}^2$, where $\dot{x}$ is the velocity of the mirror.

*   **Potential Energy ($V$):** The potential energy stored in the spring is $V = \frac{1}{2}kx^2$, where $x$ is the displacement of the mirror from its equilibrium position.

*   **Rayleigh Dissipation Function ($R$):** For a damped system, we introduce the Rayleigh dissipation function, which accounts for non-conservative forces like damping. For a dashpot, it is given by $R = \frac{1}{2}b\dot{x}^2$.

*   **Generalized Force ($Q$):** The external force acting on the mirror is $F(t)$. This is the generalized force in our system.


### Lagrange's Equations

Lagrange's equations for a system with non-conservative forces are given by:

$$\frac{d}{dt}\left(\frac{\partial L}{\partial\dot{x}}\right) - \frac{\partial L}{\partial x} + \frac{\partial R}{\partial\dot{x}} = Q$$

where $L = T - V$ is the Lagrangian.


### Derivations

First, we compute the Lagrangian $L = T - V$:

$$L = \frac{1}{2}m\dot{x}^2 - \frac{1}{2}kx^2$$

Next, we calculate the partial derivatives needed for Lagrange's equations:

1.  **$\frac{\partial L}{\partial\dot{x}}$:**
    $$\frac{\partial L}{\partial\dot{x}} = m\dot{x}$$

2.  **$\frac{d}{dt}\left(\frac{\partial L}{\partial\dot{x}}\right)$:**
    $$\frac{d}{dt}\left(m\dot{x}\right) = m\ddot{x}$$

3.  **$\frac{\partial L}{\partial x}$:**
    $$\frac{\partial L}{\partial x} = -kx$$

4.  **$\frac{\partial R}{\partial\dot{x}}$:**
    $$\frac{\partial R}{\partial\dot{x}} = b\dot{x}$$

Substituting these into Lagrange's equations, we get:

$$m\ddot{x} - (-kx) + b\dot{x} = F(t)$$

Rearranging the terms, we obtain the equation of motion for the mirror:

$$m\ddot{x} + b\dot{x} + kx = F(t)$$

This is the familiar equation for a damped, driven harmonic oscillator, which accurately describes the motion of the LIGO mirror under the influence of a gravitational wave. We will solve this numerically for a specific $F(t)$ corresponding to a passing gravitational wave.

## Numerical Solution

We will now numerically solve the equation of motion derived above:

$$m\ddot{x} + b\dot{x} + kx = F(t)$$

To solve this second-order ordinary differential equation (ODE) numerically, we convert it into a system of two first-order ODEs. Let $y_0 = x$ and $y_1 = \dot{x}$. Then we have:

1.  $\dot{y_0} = y_1$
2.  $\dot{y_1} = \frac{F(t) - b\dot{x} - kx}{m} = \frac{F(t) - by_1 - ky_0}{m}$

So, our system of first-order ODEs is:

$$\begin{pmatrix} \dot{y_0} \\ \dot{y_1} \end{pmatrix} = \begin{pmatrix} y_1 \\ \frac{F(t) - by_1 - ky_0}{m} \end{pmatrix}$$

We will use `scipy.integrate.odeint` to solve this system. The external force $F(t)$ will be modeled as a Gaussian pulse, representing a short burst of gravitational radiation. For simplicity, we choose a Gaussian pulse centered at $t=5$ seconds with a width of $\sigma = 0.1$ and an amplitude of $A = 10^{-20}$ Newtons. We also use realistic parameters for the LIGO mirror system:

*   **Mass ($m$):** $40 \text{ kg}$ (standard LIGO mirror mass)
*   **Spring Constant ($k$):** $1 \text{ N/m}$ (representing the suspension system's stiffness)
*   **Damping Coefficient ($b$):** $0.1 \text{ Ns/m}$ (representing damping in the system)

These parameters are chosen to illustrate the mirror's response to a very weak gravitational wave signal.

We will integrate from $t=0$ to $t=10$ seconds with initial conditions $x(0) = 0$ and $\dot{x}(0) = 0$ (mirror starts at rest at equilibrium).

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the system parameters
m = 40  # kg (mass of the mirror)
k = 1   # N/m (spring constant)
b = 0.1 # Ns/m (damping coefficient)

# Define the external force F(t) - a Gaussian pulse
# This simulates a short burst of gravitational radiation
def F_t(t, A=1e-20, t_center=5, sigma=0.1):
    return A * np.exp(-(t - t_center)**2 / (2 * sigma**2))

# Define the system of first-order ODEs
def model(y, t, m, b, k):
    x, v = y  # y[0] = x (position), y[1] = v (velocity)
    dydt = [
        v,
        (F_t(t) - b * v - k * x) / m
    ]
    return dydt

# Initial conditions
y0 = [0, 0]  # x(0) = 0, v(0) = 0

# Time points for the simulation
t = np.linspace(0, 10, 1000)  # Simulate from t=0 to t=10 seconds

# Solve the ODE
sol = odeint(model, y0, t, args=(m, b, k))

# Extract position and velocity
x_solution = sol[:, 0]
v_solution = sol[:, 1]

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(t, x_solution * 1e20) # Scale by 1e20 for better visualization of tiny displacement
plt.title('Displacement of LIGO Mirror due to Gravitational Wave')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (x10^-20 m)')
plt.grid(True)
plt.show()
```

## Results and Discussion

The plot above shows the displacement of the LIGO mirror over time due to the simulated gravitational wave pulse. As expected, the displacement is extremely small, on the order of $10^{-20}$ meters, which is consistent with the incredibly small strains caused by gravitational waves. The mirror responds to the Gaussian pulse, showing a temporary displacement that then decays due to the damping present in the system.

### Key Observations:

*   **Tiny Displacements:** The maximum displacement is approximately $4 \times 10^{-20}$ meters, highlighting the extreme sensitivity required for gravitational wave detection.
*   **Damped Oscillation:** After the pulse, the mirror's displacement returns to zero, exhibiting a damped oscillatory behavior as the energy dissipates due to the dashpot.
*   **Response to Force:** The displacement curve closely follows the shape of the applied Gaussian force, demonstrating the direct relationship between the gravitational wave and the mirror's motion.

This simple model, while not capturing the full complexity of a multi-dimensional interferometer, effectively demonstrates how Lagrange's equations can be used to model the fundamental dynamics of a crucial component in gravitational wave detectors. The numerical solution confirms the expected physical behavior of a damped, driven harmonic oscillator under the influence of a transient external force.
